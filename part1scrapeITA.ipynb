```python
# request gets the data.  soup structures the data.
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
requests.urllib3.disable_warnings()
```

```python
def process_player():
    #soup.find('div', {'id': 'overall'})
    #soup.find_all('td')
    #soup.find_all('table')[3]
    #soup.find_all('table')[3].find_all('td')[0]
    #soup.find_all('table')[3].find_all('td')[7]
    #soup.find_all('table')[3].find_all('td')[14]
    #len(soup.find_all('table')[3].find_all('td'))

    current_player=str(soup.find_all('td')[13].find_all('option', selected=True)).split('</option>')[-2].split('>')[-1]
    current_player_school=str(soup.find_all('td')[11].find_all('option', selected=True)).split('</option>')[-2].split('>')[-1]
    current_player_school=current_player_school.replace('&amp;','&')
    current_season=str(soup.find_all('td')[17].find_all('option', selected=True)).split('</option>')[-2].split('>')[-1]

    tag_cnt=len(soup.find_all('table')[3].find_all('td'))

    date_column=[]
    event_column=[]
    round_column=[]
    opponent_column=[]
    opp_school_column=[]
    result_column=[]
    score_column=[]

    idx=0
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        #my_output=str(my_output).split('\r\n')[1].strip()
        my_output=str(my_output).split('\n')[1].strip()
        my_output=my_output.replace('</td>', '')
        idx=idx+7
        date_column.append(my_output)
        #print(my_output)

    idx=1
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        my_output=str(my_output).split('</font></a>')[0].strip()
        my_output=my_output.split('>')[-1].strip()
        idx=idx+7
        event_column.append(my_output)
        #print(my_output)

    idx=2
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        #my_output=str(my_output).split('\r\n')[1].strip()
        my_output=str(my_output).split('\n')[1].strip()
        my_output=my_output.replace('</td>', '')
        idx=idx+7
        round_column.append(my_output)
        #print(my_output)

    idx=3
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        my_output=str(my_output).split('</a>')[0].strip()
        my_output=my_output.split('>')[-1].strip()
        idx=idx+7
        opponent_column.append(my_output)
        #print(my_output)

    idx=4
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        my_output=str(my_output).split('</a>')[0].strip()
        my_output=my_output.split('>')[-1].strip()
        my_output=my_output.replace('&amp;','&')
        idx=idx+7
        opp_school_column.append(my_output)
        #print(my_output)

    idx=5
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        #my_output=str(my_output).split('\r\n')[1].strip()
        my_output=str(my_output).split('\n')[1].strip()
        my_output=my_output.replace('</td>', '')
        idx=idx+7
        result_column.append(my_output)
        #print(my_output)

    idx=6
    while idx<tag_cnt:
        my_output=soup.find_all('table')[3].find_all('td')[idx]
        #my_output=str(my_output).split('\r\n')[1].strip()
        my_output=str(my_output).split('\n')[1].strip()
        my_output=my_output.replace('</td>', '')
        idx=idx+7
        score_column.append(my_output)
        #print(my_output)

    player_results = pd.DataFrame(
        {'date': date_column[1:],
         'event': event_column[1:],
         'round': round_column[1:],
         'opponent':opponent_column[1:],
         'opponent school':opp_school_column[1:],
         'result':result_column[1:],
         'score':score_column[1:]
        })

    player_results.insert(0, 'current player', current_player)
    player_results.insert(1, 'current player school', current_player_school)
    player_results.insert(2, 'current season', current_season)
    return player_results
```

```python
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=18&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=17&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=16&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=15&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=14&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=13&PlayerId=0&TypeId=1'
base_team_url='http://itarankings.itatennis.com/PlayerSchedule.aspx?did=1&confid=0&teamid=0&Seasonid=12&PlayerId=0&TypeId=1'
response = requests.get(base_team_url, verify=False)
response
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
final_stats = pd.DataFrame(
    {'current player':[],'current player school':[],'current season':[],'date':[],
     'event':[],'round':[],'opponent':[],'opponent school':[],'result':[],'score':[]})

# get all teams... and loop through them
soup.find_all('td')[11]
for p in soup.find_all('td')[11].find_all('option')[0:2]:
#for p in soup.find_all('td')[11].find_all('option'):
    my_team=str(p).split('</option>')[-2].split('>')[-1]
    my_team=my_team.replace('&amp;','&')
    team_id=str(p).split('</option>')[0].split('>')[-2].split('=')[-1].replace('"', '')
    #print(my_team,": ",team_id)
    team_url=base_team_url.replace("teamid=0","teamid="+team_id)
    #print(team_url)
    
    response = requests.get(team_url, verify=False)
    response
    soup = BeautifulSoup(response.text, 'html.parser')
    
    for p in soup.find_all('td')[13].find_all('option'):
        my_player=str(p).split('</option>')[-2].split('>')[-1]
        if my_player!='Select Player...':
            player_id=str(p).split('</option>')[0].split('>')[-2].split('=')[-1].replace('"', '')
            #print("player id: ",player_id,"; ",my_player)
            player_url=team_url.replace("PlayerId=0","PlayerID="+player_id)

            response = requests.get(player_url, verify=False)
            response
            soup = BeautifulSoup(response.text, 'html.parser')
            table_rows=len(soup.find_all('table')[3])
            #print(my_player,"; ",table_rows)
            # if len=3, then no table... skip this player
            if table_rows>3:
                final_stats=pd.concat([final_stats,process_player()],sort=False)
```

```python
final_stats.to_csv('./output/my_output_2012_part5.csv', encoding='utf-8', index=False)
```
